```python
class Sandwich(object):
    def __init__(self,owner, bread="white"):
        self.owner = owner
        self.bread = bread
        self.toppings = []

    def __add__(self, other):
        if isinstance(other,str):
            self.toppings.append(other)
        else:
            raise RuntimeError("Toppings Must be Strings")

    def __sub__(self, other):
        if isinstance(other,str):
            while other in self.toppings:#if other in self.toppings:
                self.toppings.remove(other)
        else:
            raise RuntimeError("Toppings Must be Strings")

    def __eq__(self,other):
        same_bread = (self.bread == other.bread)
        same_toppings = (sorted(self.toppings)==sorted(other.toppings))
        if same_bread & same_toppings :
            return True
        else:
            return False

    def __neq__(self,other):
        return not self==other

    def __repr__(self):
        rep =f"""
            SANDWICH
            Owner:{self.owner}
            Bread:{self.bread}
            Toppings:{", ".join(self.toppings)}
            price:{self.get_price()}
             """
        return rep

    def get_price(self,discount=0):
        price = 0
        if self.bread != "white":
            price += 2
        price += len(self.toppings)
        return round(price *(100 -discount)/100,2)



```
